```python
import pandas as pd
import numpy as np
```

# Filter all data on startYear >= 2010 & titleType = Movies

## Load Title_basics
DataFrame with the information of the movies

```python
#Load dataset
title_basics = pd.read_csv('data_raw/title.basics.tsv/data.tsv', sep='\t', index_col = 'tconst', low_memory=False)
```

```python
title_basics
```

```python
title_basics.isna().sum()
```

```python
title_basics= title_basics.dropna()
title_basics.isna().sum()
```

```python
title_basics.startYear.value_counts()
```

```python
title_basics = title_basics[title_basics['startYear'] != r'\N']
title_basics = title_basics[title_basics['genres'] != r'\N']
```

```python
title_basics.startYear.value_counts()
```

```python
title_basics['startYear'] = title_basics['startYear'].astype(int)
```

```python
title_basics.dtypes
```

```python
title_basics = title_basics\
                .loc[(title_basics['startYear'] >= 2010) & (title_basics['titleType'] == 'movie')]\
                .sort_values(by=['startYear'])
    
title_basics
```

## Load title_ratings
Merge rating DataFrame with the movies DataFrame

```python
title_ratings = pd.read_csv('data_raw/title.ratings.tsv/data.tsv', sep='\t', index_col = 'tconst')
title_ratings
```

```python
title_basics = pd.merge(title_basics,title_ratings, on = 'tconst', how='inner')
title_basics
```

```python
del title_ratings
```

## Load title_crew
Filter the DataFrame with the movies crew(directors and writers) on the movies that are in the movies DataFrame

```python
title_crew = pd.read_csv('data_raw/title.crew.tsv/data.tsv', sep='\t', index_col = 'tconst')
```

```python
title_crew
```

```python
title_crew = title_crew.loc[title_crew.index.isin(title_basics.index)]
```

```python
title_crew
```

## Load title_principals
Filter the DataFrame with the movies principals(actors, actress, director, writer and producer) on the movies that are in the movies DataFrame

```python
title_principals = pd.read_csv('data_raw/title.principals.tsv/data.tsv', sep='\t')
```

```python
title_principals
```

```python
title_principals = title_principals.loc[title_principals['tconst'].isin(title_basics.index)]
title_principals = title_principals[title_principals['category'].isin(['actor','actress'])]
title_principals = pd.merge(title_principals,title_basics[['startYear','averageRating']], on = 'tconst')
```

```python
title_principals
```

## Load name_basics
Filter the DataFrame with the crew names on the DataFrame with movies

```python
name_basics = pd.read_csv('data_raw/name.basics.tsv/data.tsv', sep='\t', index_col = 'nconst')
```

```python
name_basics
```

```python
name_basics = name_basics.loc[name_basics.index.isin(title_principals['nconst'])]
```

```python
name_basics
```

## Create dummy variables for genres on the title_basics dataframe

```python
title_basics = pd.merge(title_basics.drop('genres', axis = 1),title_basics['genres'].str.get_dummies(sep = ','), on = 'tconst')
```

```python
title_basics.head()
```

# Export Filtered dataframes to csv

```python
title_basics.to_csv('data/movies_info.csv')
title_crew.to_csv('data/movies_crew.csv')
title_principals.to_csv('data/movies_actors.csv', index=False)
name_basics.to_csv('data/crew_names.csv')
```
